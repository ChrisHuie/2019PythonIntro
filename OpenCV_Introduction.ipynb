```
#create 300x300 numpy array and 250x250 yellow rectangle inside it
rectangle = np.zeros((300,300), dtype="uint8")
cv2.rectangle(rectangle, (25,25), (275,275), 255, -1)
plt.imshow(rectangle)
```

```
#create 300x300 numpy array and 150 radius yellow circle inside it
circle = np.zeros((300,300), dtype="uint8")
cv2.circle(circle, (150,150), 150, 255, -1)
plt.imshow(circle)
```

```
#performing the bitwise_and operation
bitwiseAnd = cv2.bitwise_and(circle, rectangle)
plt.imshow(bitwiseAnd)
```

```
#performing the bitwise_or operation
bitwiseOr = cv2.bitwise_or(rectangle, circle)
plt.imshow(bitwiseOr)
```

```
#performing the bitwise_xor operation - only when 1 input is true | colors inverse of bitwise_and
bitwiseXor = cv2.bitwise_xor(rectangle, circle)
plt.imshow(bitwiseXor)
```

```
#create numpy array filled with zeros with exact width and height of image
#create a square mask
mask = np.zeros(img.shape[:2], dtype="uint8")
(cX, cY) = (img.shape[1]//2, img.shape[0]//2)
cv2.rectangle(mask, (cX - 300, cY- 300), (cX + 300, cY + 300), 255, -1)
plt.imshow(mask)
```

```
masked = cv2.bitwise_and(img, img, mask=mask)
plt.imshow(masked)
```

```
coins = cv2.imread('/content/coins.jpg')

#convert image to grayscale
img = cv2.cvtColor(coins, cv2.COLOR_BGR2GRAY)

#implement gaussian blur
blurred = cv2.GaussianBlur(img, (5,5), 0)
# plt.imshow(blurred)

#simple thresholding using binary
(T, thresh) = cv2.threshold(blurred, 155, 255, cv2.THRESH_BINARY)
# plt.imshow(thresh)

#simple thresholding using inv binary
(T, threshInv) = cv2.threshold(blurred, 155, 255, cv2.THRESH_BINARY_INV)
# plt.imshow(threshInv)


#plt.imshow( cv2.bitwise_and(img, img, mask=threshInv))



#adaptive thresholding using mean
thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 4)
#plt.imshow(thresh)

#adaptive thresholding using gaussian
threshGaussian = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 4)
plt.imshow(threshGaussian)
```

```
#convert image to grayscale
gray = cv2.cvtColor(coins, cv2.COLOR_BGR2GRAY)

#implement gaussian blur
blurred = cv2.GaussianBlur(gray, (15,15), 0)
# plt.imshow(blurred)

canny = cv2.Canny(blurred, 30, 150)
#plt.imshow(canny)

#finding the contours, counting, and marking them
(cont , cnts, mark) = cv2.findContours(canny.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print("The number of coins in the image is: {}".format(len(cnts)) )

#create a copy of the image
coinImg = coins.copy()

#draw the contours in the actual color image copy
cv2.drawContours(coins, cnts, -1, (0, 255, 0), 2)
plt.imshow(coinImg)
```
